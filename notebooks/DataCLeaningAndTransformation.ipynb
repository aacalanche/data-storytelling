```python
import pandas as pd
import numpy as np

# 1. Create & inspect the DataFrame
data2 = {
    "city": ["NY", "LA", "NY", "SF", "LA", "NY"],
    "store_id": [1, 2, 1, 3, 2, 1],
    "sales": ["100.0", "200.5", None, "300", "250.0", "120.0"],
    "customers": [10, 20, 15, None, 22, 12],
}
stores = pd.DataFrame(data2)

print("=== 1. INSPECT ===")
print(stores)
print(stores.dtypes)
print(stores.isna().sum())

# 2. Handle missing values
stores['sales'] = pd.to_numeric(stores['sales'], errors='coerce')
sales_mean = stores['sales'].mean()  # ~194.1
stores['sales'] = stores['sales'].fillna(sales_mean)
stores = stores.dropna(subset=['customers'])  # Drop SF row

# 3. Convert types
stores['city'] = stores['city'].astype('category')
stores['store_id'] = stores['store_id'].astype('int32')

print("\n=== 2-3. CLEANED ===")
print(stores)
print(stores.dtypes)

# 4. Grouped aggregation
grouped = stores.groupby('city').agg({
    'sales': 'sum',
    'customers': 'mean'
}).round(2)

print("\n=== 4. GROUPED AGGREGATION ===")
print(grouped)

# 5. Pivot table (handles duplicate NY store 1)
pivot = stores.pivot_table(
    index='city',
    columns='store_id',
    values='sales',
    aggfunc='sum'
)

print("\n=== 5. PIVOT TABLE ===")
print(pivot)
```
